```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
```

```python
model = keras.Sequential([keras.layers.Dense(units = 1, input_shape = [1])])

'''
# Every neuron is connected to every other neuron, which is called a dense layer \n\
because of the dense connectivity between them.

# Keras is a framework within TensorFlow for making the definition of training and \n\
using ML models easily.

# Sequential tells that we're going to define a sequence of layers.
'''

model.compile(optimizer = 'sgd', loss = 'mean_squared_error')

'''
# Loss function is a way of measuring how good or bad the guess is, and the results \n\
of this are then used by optimizer to make another guess. That guess is then measured \n\
using the loss function again with the logic being that an optimizer can improve the guess.
'''

xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype = float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype = float)

model.fit(xs, ys, epochs = 500)
```

```python
print(model.predict([10.0]))
```
